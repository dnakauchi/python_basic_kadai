```python
class Human:

    def __init__(self,name,age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >=20:
            print(self.name+" is adult.")
        else:
            print(self.name+" is not adult.")

personal_data = {"Taro": 40, "Jiro": 36, "Saburo": 32, "shiro": 18, "Goro": 12}

for key, value in personal_data.items():
  human = Human(key,value)
  human.check_adult()
```
